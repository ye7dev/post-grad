### 표본의 크기와 표본 통계량의 분포 

```r
# 평균구하는 거 없이 샘플만 1000개 = 데이터 분포 
samp_data <- data.frame(income=sample(loans_income, size=1000), type = 'data_dist')
head(samp_data)
```

```r
# 5개 값이 하나의 표본평균 구하는 데 사용됨 -> 1000개의 표본평균이 표본 분포를 이룬다 (!= 5000개의 데이터가 데이터 분포를 이룬다) 
samp_mean_05 <- data.frame(income=tapply(sample(loans_income, 1000*5), rep(1:1000, rep(5, 1000)), FUN=mean), type='mean_of_5')
```

###### rep 함수 실행

```r
rep(5, 1000)
```

```r
length(rep(5, 1000))
```

```r
length(rep(1:1000, rep(5, 1000)))
```

```r
# 1:1000 -> 1, ..., 1000 = sequence (길이 1000)
# rep(5, 1000) -> 5, ... 5  = times (길이 1000)
## sequence[i]를 times[i]만큼 반복한다 
rep(1:1000, rep(5, 1000)) 
```

###### done

```r
# 20개 값이 하나의 표본평균
samp_mean_20 <- data.frame(income=tapply(sample(loans_income, 1000*20), rep(1:1000, rep(20, 1000)), FUN=mean), type='mean_of_20')
```

```r
# 세 df를 위아래로 쌓는다 (행 단위로 결합)
income <- rbind(samp_data, samp_mean_05, samp_mean_20)
income$type = factor(income$type, levels=c('data_dist', 'mean_of_5', 'mean_of_20'), labels=c('Data', 'Mean of 5', 'Mean of 20'))
```

```r
head(income)
```

```r
ggplot(income, aes(x=income)) + geom_histogram(bins=40) + facet_grid(type ~.)
```

### QQ plot

```r
# row.names=1: first col is row name
sp500_px <- read.csv(file.path(PSDS_PATH, 'data', 'sp500_data.csv.gz'), row.names=1)
```

```r
# 일일 주식 수익률 (로그 변환된 주식 가격의 차분)
nflx <- sp500_px[, 'NFLX']
nflx <- diff(log(nflx[nflx>0]))
qqnorm(nflx)
abline(a=0, b=1, col='grey')
```

### Various Distributions

```r
# binomial
dbinom(x=2, size=5, p=0.1) #pmf, 정확히 성공이 x번 나올 확률
```

```r
pbinom(2, 5, 0.1) #cdf, 성공이 x번 이하로 나올 확률
```

```r
# poisson
les_poissons = rpois(100, lambda=2) # 포아송 분포를 따르는 난수를 몇 개 생성할 것인지, 람다 값은 얼마인지
les_poissons
```

```r
mean(les_poissons)
```

```r
var(les_poissons)
```

```r
# 생성하는 난수의 개수가 커질 수록 목표하는 평균, 분산 값에 가깝게 된다 
plus_poissons = rpois(1000, lambda=2)
cat(mean(plus_poissons), var(plus_poissons))
```

```r
# exponential
rexp(n=100, rate=0.2)
```

```r
# weibull
rweibull(100, 1.5, 5000) # 발생 개수, shape(발생률 변화), scale(특성 수명)
```
